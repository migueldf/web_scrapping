```python
import time
import re
import sys
import requests
import numpy as np
import pandas as pd
from pathlib import Path

from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
url_linkedin = 'https://www.linkedin.com/jobs/search/?keywords=Data%20Analyst&location=Mexico%20City%20Metropolitan%20Area&geoId=90010045&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'

driver = webdriver.Chrome()

driver.get(url_linkedin)
```

ActionChains(driver).send_keys(Keys.END).perform()
time.sleep(3)

```python
page_source = driver.page_source

driver.quit()
```

```python
linkedin = BeautifulSoup(page_source,'lxml')

print(linkedin.prettify())
```

```python
linkedin_links = linkedin.select('a.disabled ember-view job-card-container__link job-card-list__title')

linkedin_links = [str(link['href']) for link in linkedin_links]

linkedin_links
```
